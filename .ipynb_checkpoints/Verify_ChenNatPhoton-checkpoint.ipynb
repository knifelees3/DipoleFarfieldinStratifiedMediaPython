```python
import numpy as np
import datetime
from class_BFP_Image_QD import BFP_Image_QD
from scipy.optimize import curve_fit
import Fun_BFP_Image
import matplotlib.pyplot as plt
```

## Define the basci parameters

### Structure 1

```python
WL0 = 580e-9
nUp = 1
nDn = 1

# Initialize the coordinate for each layer
# The dipole should be better in the coordinate z=0
dl = np.zeros((nUp + nDn, 1))
dis = 400e-9
dl[1] = 600e-9
dl[0] = 0

# The position of the dipole
POSD = dl[nDn - 1] + dis

# Dipole direction
# phi = np.pi / 2 + np.pi / 6
# dx = np.sin(16 / 180 * np.pi) * np.cos(phi)
# dy = np.sin(16 / 180 * np.pi) * np.sin(phi)
# dz = np.cos(16 / 180 * np.pi)
# p0 = np.array([dx, dy, dz])
p0 = np.array([0, 0, 1])

# Initialize the permitivity
Eplist = np.zeros((nUp + nDn + 1, 1), dtype=complex)


Eplist[2] = 1.78**2
Eplist[1] = 1.5**2
Eplist[0] = 1

D1 = BFP_Image_QD(Eplist, dl, nUp, nDn, p0, WL0, POSD)

num_theta =200 
num_phi = 200

NA=1.78
theta_mat=np.linspace(0,np.pi/2,num_theta)
phi_mat=np.linspace(0,np.pi*2,num_phi)
theta_grid,phi_grid=np.meshgrid(theta_mat,phi_mat)


kx_grid = np.sin(theta_grid)*np.cos(phi_grid)*D1.k0*NA
ky_grid = np.sin(theta_grid)*np.sin(phi_grid)*D1.k0*NA
```

We can visualize the structure now

```python
D1.Show_Structure()
```

```python
# Initialize the Green Function
D1.Cal_Green_List(kx_grid, ky_grid)
```

```python
# To calculate and plot the pattern for a given p0
PatternUp1,PatternDn1 = D1.Cal_Pattern_List_QD_p1(p0)
```

```python
fig1 = plt.figure()
plt.pcolormesh(kx_grid,ky_grid,PatternUp1, cmap='jet')
plt.savefig("./Figures/VerifyChenStr12D.png")
plt.show()
```

```python
nPat1_rho=np.sum(PatternUp1*krho_grid,axis=0)
```

```python
fig2=plt.figure()
plt.plot(theta_mat/np.pi*180,nPat1_rho)
plt.savefig("./Figures/VerifyChenStr1DRho.png")
plt.show()
```
